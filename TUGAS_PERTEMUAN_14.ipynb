```
import numpy as np
import matplotlib.pyplot as plt
```

Fungsi

```
def cos(frek,tipe,ht,hr,d):
  #frek = frekuensi (MHz)
  #tipe = Urban, SubUrban, Open
  #ht = tinggi tower (m)
  #hr = tinggi user (m)
  #d = jarak (m)
    if frek in range(150,1501):
        Ca=69.55
        Cb=26.16
    elif frek in range(1501,2001):
        Ca=46.3
        Cb=33.9
    
    if tipe=='urban' :
        Cm=0
    elif tipe=='suburban' :
        Cm= -2*((np.log10((frek/28)**2))-5.4)
    elif tipe=='open' :
        Cm= -4.78*(np.log10(frek))+18.33*(np.log10(frek))-40.94
    else :
        Cm='tipe salah'
    
    ahr = (1.1*np.log10(frek)-0.7)*hr-(1.56*np.log10(frek)-1.8)
    Lp = Ca+(Cb*np.log10(frek))-(13.83*np.log(ht))-ahr+((44.9-6.55*np.log10(ht))*np.log10(d))-Cm
    
    return Lp
```

```
#nilai daya terima
def daya_terima(ptx,Lp, gtx, grx):
  prx = 10*np.log10(ptx) - Lp + gtx + grx
  
  return 10**(prx/10)

#nili signal to noise ratio
def s2nr (bw, prx):
  no = -174 + 10*np.log10(bw)
  no1 = 10**(no/10)
  
  return prx/no1

#kapasitas seluler
def kapasitas_seluler (bw, snr):
  ks= bw * np.log2(1+snr)

  return ks
```

```
ht = 80 #m
hr = 1.5 #m
tipe = ['urban', 'suburban', 'open']
fr = [ 700, 900, 1800]
jmax = 1000 #m
jmin = 300 #m
titik = 200 
d = np.random.randint(jmin, jmax, titik)
d = d[np.argsort(d)]
ptx = 30 #Watt
gtx = 10 #dB
grx = 2 #dB
bw = 5000000 #Hz
```

Perhitungan Pada Frekuensi 1800 MHz

```
lp1 = []
prx1 = []
snr1 = []
ks1 = []
for i in range(3):
  lp1.append(cos(fr[2], tipe[i], ht, hr, d))

  prx1.append(daya_terima(ptx, lp1[i], gtx, grx))
  prx_sort = prx1[i]
  prx1[i] = prx_sort[np.argsort(d)]

  snr1.append(s2nr(bw, prx1[i]))
  snr_sort = snr1[i]
  snr1[i]= snr_sort[np.argsort(d)]

  ks1.append(kapasitas_seluler(bw, snr1[i]))
  ks_sort = ks1[i]
  ks1[i]= ks_sort[np.argsort(d)]

  #grafik daya terima pada frekuensi 1800MHz dgn 3 tipe area (Urban,SubUrban,Open)

plt.plot(d, prx1[0], 'g')
plt.plot(d, prx1[1], 'r')
plt.plot(d, prx1[2], 'm')
plt.title('Grafik Daya Terima Pada Frekuensi 1800MHz Dgn 3 Tipe Area (Urban,SubUrban,Open)')
plt.ylabel('Daya Terima (milliWatt)')
plt.xlabel('Jarak (m)')
plt.legend(['urban', 'Suburban', 'Open'], loc='upper right')
plt.show()

  #grafik signal to noise ratio pada frekuensi 1800MHz dgn 3 tipe area (Urban,SubUrban,Open)

plt.plot(d, snr1[0], 'c')
plt.plot(d, snr1[1], 'r')
plt.plot(d, snr1[2], 'g')
plt.title('Grafik Signal to Noise Ratio Dgn Frekuensi 1800MHz di 3 Tipe Area (Urban,SubUrban,Open)')
plt.ylabel('Signal to Noise Ratio')
plt.xlabel('Jarak (m)')
plt.legend(['Urban', 'Suburban', 'Open'], loc='upper right')
plt.show()

  #grafik kapasitas seluler dgn frekuensi 1800MHz di 3 tipe area (Urban,SubUrban,Open)

plt.plot(d, ks1[0], 'c')
plt.plot(d, ks1[1], 'r')
plt.plot(d, ks1[2], 'b')
plt.title('Grafik Kapasitas Seluler Dgn Frekuensi 1800MHz di 3 Tipe Area (Urban,SubUrban,Open)')
plt.ylabel('Kapasitas (bit per second)')
plt.xlabel('Jarak (m)')
plt.legend(['Urban', 'Suburban', 'Open'], loc='upper right')
plt.show()
```
